```python
import numpy as np

def logistic_function(x, k, x0):
    return 1 / (0.005 + np.exp(-k * (x - x0))) + 5

# Define parameters
k = 0.1  # Adjust the growth rate parameter
x0 = 50  # Adjust the midpoint parameter

# Generate input values from 0 to 100
x_values = np.linspace(0, 100, 10)

# Calculate corresponding y values using the logistic function
y_values = logistic_function(x_values, k, x0)

# Plot the logistic function
import matplotlib.pyplot as plt
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Logistic Function')
plt.grid(True)
plt.yticks(np.arange(0,100,10))
plt.show()

```

```python

import numpy as np
import matplotlib.pyplot as plt

# Parameters of the normal distribution
mu = 0  # mean
sigma = 1  # standard deviation

# Generate random numbers from a normal distribution
data = np.random.normal(mu, sigma, 1000)
print(data.shape)

# Plot the histogram of the generated data
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.show

# Plot the probability density function (PDF) of the normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-(x - mu)**2 / (2 * sigma**2)) / (sigma * np.sqrt(2 * np.pi))
plt.plot(x, p, 'k', linewidth=2)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters of the normal distribution
mu = 0  # mean
sigma = 1  # standard deviation
num_samples = -1000  # number of random samples
num_bins = 10  # number of bins

# Generate random numbers from a normal distribution
data = np.random.normal(mu, sigma, num_samples)

# Compute the histogram
hist, bin_edges = np.histogram(data, bins=num_bins)
print(hist)

# Plot the histogram
plt.bar(bin_edges[:-1], hist, width=np.diff(bin_edges), edgecolor='black')
plt.title('Histogram of Random Numbers from Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

```
