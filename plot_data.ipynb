```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("imm_dep_data.csv")[:10]
```

```python
X = data.iloc[:, 0].astype(float) / 10000
y2042 = data.iloc[:, 1]
y2062 = data.iloc[:, 2]

plt.plot(X, y2042, label="2042")
plt.legend()
plt.plot(X, y2062, label="2062")
plt.legend()
plt.xlabel("Immigration per year since 2022 (in 10,000)")
plt.ylabel("Dependency ratio ")
```

```python
data = pd.read_csv("imm_tfr_dep_data.csv")
immigration = data.iloc[:, 0]
tfr = data.iloc[:, 1]
dep_ratio = data.iloc[:, 2]
population = data.iloc[:, 3]
```

```python
dep_ratio.values.reshape(6, 6)
```

```python
import seaborn as sns

plt.figure(figsize=(10, 7))
sns.heatmap(dep_ratio.values.reshape(6, 6), annot=True, cmap="coolwarm", xticklabels=tfr.unique(), yticklabels=immigration.unique())
plt.xlabel("TFR")
plt.ylabel("Immigration per year")
```

```python
plt.figure(figsize=(10, 7))
sns.heatmap(population.values.reshape(6, 6) / 1000, annot=True, cmap="coolwarm", xticklabels=tfr.unique(), yticklabels=immigration.unique())
plt.xlabel("TFR")
plt.ylabel("Immigration per year")
```
