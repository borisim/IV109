```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("imm_dep_data.csv")[:10]
```

```python
X = data.iloc[:, 0].astype(float) / 10000
y2042 = data.iloc[:, 1]
y2062 = data.iloc[:, 2]

plt.plot(X, y2042, label="2042")
plt.legend()
plt.plot(X, y2062, label="2062")
plt.legend()
plt.xlabel("Immigration per year since 2022 (in 10,000)")
plt.ylabel("Dependency ratio ")
```

```python
df = pd.read_csv("deps.csv")
df
```

```python
# Drop first column
df = df.drop(df.columns[0], axis=1)

# Turn rows into arrays
deps = df.values
deps.shape
```

```python
# compute average dependency ratio for each row

avg_deps = deps.mean(axis=1)
avg_deps

sd_deps = deps.std(axis=1)
sd_deps
```

```python
# Make x axis start at 2022

years = range(2022, 2022 + deps.shape[1])

# set figure size
plt.figure(figsize=(10, 7))

plt.plot(years, deps[0], label="Stratgia 1, DR priemer = {:.2f}, std = {:.2f}".format(avg_deps[0], sd_deps[0]))
plt.plot(years, deps[1], label="Stratgia 2, DR priemer = {:.2f}, std = {:.2f}".format(avg_deps[1], sd_deps[1]))
plt.plot(years, deps[2], label="Stratgia 3, DR priemer = {:.2f}, std = {:.2f}".format(avg_deps[2], sd_deps[2]))
plt.plot(years,deps[3], label="Stratgia 4, DR priemer = {:.2f}, std = {:.2f}".format(avg_deps[3], sd_deps[3]))
plt.plot(years,deps[4], label="Stratgia 5, DR priemer = {:.2f}, std = {:.2f}".format(avg_deps[4], sd_deps[4]))
plt.plot(years,deps[5], label="Stratgia 6, DR priemer = {:.2f}, std = {:.2f}".format(avg_deps[5], sd_deps[5]))
plt.legend()
plt.xlabel("Rok")
plt.ylabel("Dependency ratio")
```
